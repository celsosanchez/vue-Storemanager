```python
#import the library
import numpy as np
import pandas as pd
from apyori import apriori

#load the data
store_data = pd.read_csv('./data/groceries - groceries.csv', sep=',')
store_data.head()

transactions = []

# Add all the items from each row in one list( Neglect the 1st columns where all the items are in number (0-9))
for i in range(0, 9835):
    transactions.append([str(store_data.values[i,u]) for u in range(1, 33)])
```

```python
#applying apirori
association_rules = apriori(transactions, min_support=0.003, min_confidence=0.2, min_lift=4, min_length=1)
```

```python
rules = list(association_rules)
```

```python
len(rules)
```

```python
best_rules = []

for i in range(0, len(rules)):
    for j in range (0, len(rules[i][2])):
        best_rules.append({"lh": rules[i][2][j].items_base,"rh": rules[i][2][j].items_add, "lift":rules[i][2][j].lift })
len(best_rules)
```

```python
best_rules[0]["lift"]

def takeLift(elem):
        return elem["lift"]

sorted_rules = sorted(best_rules,key=takeLift, reverse=True)
```

```python
df = pd.DataFrame(sorted_rules)
```
